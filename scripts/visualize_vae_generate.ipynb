```python
import numpy
input_0 = numpy.load('vae_generate/input_000000.npy')
recons_0 = numpy.load('vae_generate/recons_000000.npy')
input_0.shape, recons_0.shape
```

```python
import matplotlib.pyplot as plt
#Visualize skeleton
def visualize_skeleton(skeleton):
    # shape of skeleton is 116 x 2
    left_and_right_arm = skeleton[0:4,:]
    left_hand = skeleton[4:25,:]
    right_hand = skeleton[25:46,:]
    face = skeleton[46:116,:]
    plt.scatter(left_and_right_arm[:,0],-left_and_right_arm[:,1])
    plt.scatter(left_hand[:,0],-left_hand[:,1])
    plt.scatter(right_hand[:,0],-right_hand[:,1])
    plt.scatter(face[:,0],-face[:,1])
    plt.show()
visualize_skeleton(input_0[0][2])
```

```python
visualize_skeleton(recons_0[0][2])
```
